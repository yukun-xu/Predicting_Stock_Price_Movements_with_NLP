```python
import urllib
import csv
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
StockName = 'AAPL'
DateRange = pd.date_range(start='1/1/2017', end='3/1/2017')
TopN = 5
```

```python
data = {}
for ndate in DateRange:
    data[ndate] = []
    print(ndate.strftime('%Y-%m-%d'))
    startdate = ndate.strftime('%Y-%m-%d')
    enddate = startdate
    page = urllib.request.urlopen('https://finance.google.com/finance/company_news?q=NASDAQ%3A'+StockName+
                                  '&startdate='+startdate+'&enddate='+enddate).read()
    soup= BeautifulSoup(page, 'html5lib')#.encode_contents(formatter='html')
    news = soup.find_all("div", class_="g-section news sfe-break-bottom-16")
    print ("Number of news: %d\n" % (len(news)))
    for element in news[:TopN]:
        t = element.find_all("span", class_="name")[0].get_text().replace('\xa0',' ').replace('\n','').encode('utf-8')
        print(t)
        data[ndate].append(t)
```

```python
from dateutil import parser
#scrape data from saved html files
def outputHistorical(soup_1):
    markets_1=soup_1.findAll('tbody')[1]
    date=[]
    closePrice_1=[]
    volume_1=[]
    for line in markets_1:
        #dateTemp = line.findAll('span')[0].text #date information can be included if needed
        #closePriceTemp_1 = line.findAll('span')[4].text
        """
        # Use this block of code might result in lists of different length
        try:
            closePriceTemp_1 = float(line.findAll('span')[4].get_text().replace(",", ""))
            volumeTemp_1 = float(line.findAll('span')[6].get_text().replace(",", ""))
        except:
            pass
        """
        if len(line.findAll('span'))==7: #This line is to exclude the addtion of dividend payment
            closePriceTemp_1 = float(line.findAll('span')[4].get_text().replace(",", ""))
            volumeTemp_1 = int(line.findAll('span')[6].get_text().replace(",", "")) 
            transDateTemp_1=line.findAll('span')[0].get_text()
            transDateTemp_1 = parser.parse(transDateTemp_1) #.strftime('%Y-%m-%d')
            date.append(transDateTemp_1)
            closePrice_1.append(closePriceTemp_1)
            volume_1.append(volumeTemp_1)
    return closePrice_1, volume_1, date

def reverseDate(dataframe_2):
    return dataframe_2.iloc[::-1].reset_index(drop=True)
```

```python
fileNames=["apple.html"]
closePrice=[]
volume=[]
transDate=[]
for fileName in fileNames:
    soup = BeautifulSoup(open(fileName).read())
    closePriceTemp, volumeTemp, transDateTemp=outputHistorical(soup)
    closePrice.append(closePriceTemp)
    volume.append(volumeTemp)
    transDate.append(transDateTemp)
    #print(transDate)
closePriceDF=pd.DataFrame(np.array(closePrice).T, columns=fileNames)
volumeDF=pd.DataFrame(np.array(volume).T, columns=fileNames)
transactionDF=pd.DataFrame(np.array(transDate).T, columns=fileNames)
#print(transactionDF)
reversedPriceDF=reverseDate(closePriceDF.copy())
reversedVolumeDF=reverseDate(volumeDF.copy())
reversedTransDF=reverseDate(transactionDF.copy())
#print(reversedTransDF)
```

```python
results=pd.concat([reversedTransDF, reversedPriceDF], axis=1)
#print(results.iloc[:,1].pct_change()>0)
results['return'] = (results.iloc[:,1].pct_change()>0).astype(int)
results.columns = ['Date','Price','Label']
results.index = results['Date']
results = results.drop('Date',axis=1)
print(results.head())
```

```python
tradingdate = results.index
news = []
for date in tradingdate:
    for i in range(len(df)):
        if date == df.index[i]:
            news.append(df.iloc[i-1,:].values)

news = pd.DataFrame(news)
label = pd.DataFrame(results['Label'].values)
combined_news = pd.concat([label,news],axis=1)
combined_news.index = tradingdate
combined_news.columns = ['Label','News1','News2','News3','News4','News5']
combined_news
```
